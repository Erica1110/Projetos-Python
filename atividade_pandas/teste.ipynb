```python
import pandas as pd

# Criar dados
data = {
    'Produto': ['A', 'B', 'C', 'A', 'B'],
    'Vendas': [100, 200, 150, 300, 250],
    'Região': ['Norte', 'Sul', 'Norte', 'Sul', 'Norte']
}

df = pd.DataFrame(data)

# Análise básica
print("Primeiras linhas:")
print(df.head())

print("\nVendas por produto:")
print(df.groupby(['Produto', 'Região'])['Vendas'].sum())

print("\nEstatísticas:")
print(df.describe())
```
