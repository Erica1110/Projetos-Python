```python
import pandas as pd

s = pd.Series([1, 3, 5, 7, 9])
print(s)
"""series são arrays unidimensionais que podem armazenar qualquer tipo de dados"""

```

```python
data = {
    'Nome': ['Ana', 'João', 'Maria'],
    'Idade': [25, 30, 35],
    'Cidade': ['SP', 'RJ', 'BH']
}

df = pd.DataFrame(data)
print(df)
"""dataframes são como uma tabela inteira, estrutura com linhas e colunas"""

# Primeiras linhas
df.head()

# Últimas linhas  
df.tail()

# Informações gerais
df.info()

# Estatísticas descritivas
df.describe()
```

```python
# Selecionar coluna
df['Nome']

# Selecionar múltiplas colunas
df[['Nome', 'Idade']]

# Selecionar por índice
df.iloc[0]      # Primeira linha
df.loc[[0]]       # Primeira linha (por índice)

# Filtros
df[df['Idade'] > 30]
```

```python
# Adicionar coluna
df['Salário'] = [5000, 6000, 7000]

# Renomear colunas
df.rename(columns={'Nome': 'Nomes'})

# Ordenar dados
df.sort_values('Idade', ascending=False)

# Valores únicos
df['Cidade'].unique()
# Agrupar e calcular estatísticas

print(df)
```
